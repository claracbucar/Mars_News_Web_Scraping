## Scrape Titles and Preview Text from Mars News

```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import requests
import json
import pymongo
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

### Step 1: Visit the Website

1. Use automated browsing to visit the [Mars NASA news site](https://redplanetscience.com)

```python
# Visit the Mars NASA news site
url = "https://redplanetscience.com"
browser.visit(url)
```

### Step 2: Scrape the Website

```python
# Create a Beautiful Soup object
response = browser.html
marssoup = soup(response, 'html.parser')
```

```python
# Extract all the text elements
print(marssoup.prettify())
```

```python
results = marssoup.find_all('div', class_='list_text')
```

```python
#Visualize the results
results
```

### Step 3: Store the Results

Extract the titles and preview text of the news scraped articles. Store the scraping results in Python data structures

* Store each title-and-preview pair in a Python dictionary. And, give each dictionary two keys: `title` and `preview`

* Store all the dictionaries in a Python list.

* Print the list in notebook.

```python
# Create an empty list to store the dictionaries
dict_list = []
```

```python
# Loop through the text elements
# Extract the title and preview text from the elements
# Store each title and preview pair in a dictionary
# Add the dictionary to the list

for result in results:
    title_element = result.find('div', class_='content_title')
    preview_element = result.find('div', class_='article_teaser_body')
    
    if title_element and preview_element:
        title = title_element.text.strip()
        preview = preview_element.text.strip()
        
        article_dict = {'title': title, 'preview': preview}
        dict_list.append(article_dict)
```

```python
# Print the list to confirm success
dict_list
```

```python
browser.quit()
```

### Step 4: Export the Data

Export the scraped data to a JSON file and save to Mongo Database

```python
# Export data to JSON
with open('mars.json', 'w') as f:
    json.dump(dict_list, f)
```

```python
# Export data to MongoDB
# create a MongoClient object and connect to the MongoDB server
client = pymongo.MongoClient("mongodb://localhost:27017/")

# create/connect to a database
db = client["mydatabase"]

# create/connect to a collection
collection = db["mycollection"]

# insert the list into the collection
dbresult = collection.insert_many(dict_list)

# print the inserted document IDs
print(dbresult.inserted_ids)
```
